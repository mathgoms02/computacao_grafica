```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('baby.jpg')
```

```python
B,G,R = cv2.split(image)
zeros = np.zeros_like(R)
red_images = cv2.merge([zeros,zeros,R])
green_images = cv2.merge([zeros,G,zeros])
blue_images = cv2.merge([B,zeros,zeros])
```

```python
cv2_imshow(red_images)
cv2_imshow(green_images)
cv2_imshow(blue_images)
```

```python
cv2.imwrite('/content/canal_vermelho.jpg', red_images)
cv2.imwrite('/content/canal_verde.jpg', green_images)
cv2.imwrite('/content/canal_azul.jpg', blue_images)
```
